```python
# Delete duplicate lines in q_obj.txt
with open('english-stimuli-generator\q_obj.txt', 'r') as f:
    lines = f.readlines()

# Keep only unique lines while preserving order
seen = set()
unique_lines = []
duplicates_removed = 0

for line in lines:
    if line not in seen:
        seen.add(line)
        unique_lines.append(line)
    else:
        duplicates_removed += 1

# Write unique lines back to file
with open('english-stimuli-generator\q_obj.txt', 'w') as f:
    f.writelines(unique_lines)

print(f"Total lines: {len(lines)}")
print(f"Unique lines: {len(unique_lines)}")
print(f"Duplicates removed: {duplicates_removed}")
```

```python
# Capitalize first letter, add dot, space, and <eos> to each sentence in a file
input_file = 'english-stimuli-generator/orc.txt'
output_file = 'english-stimuli-generator/orc_formatted.txt'

with open(input_file, 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]

formatted_lines = []
for line in lines:
    # Capitalize first letter
    if line:
        line = line[0].upper() + line[1:]
    # Ensure sentence ends with a dot, then add space and <eos>
    if not line.endswith('.'):
        line += '.'
    formatted_line = line + ' <eos>'
    formatted_lines.append(formatted_line)

with open(output_file, 'w', encoding='utf-8') as f:
    for line in formatted_lines:
        f.write(line + '\n')

print(f"Formatted {len(formatted_lines)} sentences. Output written to {output_file}")
```
